```python
inputs_list = base_list[0].split("\n")
operations_list = base_list[1].split("\n")

print(len(inputs_list), len(operations_list))

inputs = dict()
gates = dict()


for i in range(len(inputs_list)):
    temp = inputs_list[i].split(": ")
    inputs[temp[0]] = temp[1]
    
for i in range(len(operations_list)):
    temp1 = operations_list[i].split(" -> ")
    gates[temp1[1]] = temp1[0]


z= dict()
for gate in gates:
    if gate[0] == "z":
        z.update({gate:gates[gate]})

for i in z:
    gates.pop(i)
    
print(len(inputs),len(gates), len(z))
```

```python
def process_gates(input_dict):
    for x in input_dict:
        string = input_dict[x]
        opr_split = string.split()
        inp1, act, inp2 = opr_split
        
        
        
        if inp1 in inputs.keys() and inp2 in inputs.keys(): 
            wire1 = int(inputs.get(inp1))
            wire2 = int(inputs.get(inp2))
            result = calculate(wire1, wire2 ,act)
            inputs.update({x:result})
            calculated.update({x:input_dict[x]})
        else:
            needs_ip.update({x:input_dict[x]})
        
    if x[0] != 'z':
        for y in calculated:
            input_dict.pop(y)
      
    return
```

```python
def calculate(var1, var2, oper):
    if oper == 'XOR':
        return var1 ^ var2
    elif oper == 'AND':
        return var1 & var2
    elif oper == 'OR':
        return var1 | var2
```
