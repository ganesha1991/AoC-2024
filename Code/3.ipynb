```python
with open(r"C:\Users\thirggs\Downloads\Advent_of_code\2024\inputs\3.txt") as memory:
    # open the file, read entire file as string, strip the leading and trailing spaces
    dump = memory.read().strip()

import re
```

```python
# part 1 - filter Mul(*,*) from the list and multiply them
# regular expressions to find mul(a,b) in the string

key = r"mul\((\d+),(\d+)\)"

# findall expressions from dump 
match_list = re.findall(key,dump)

# convert list into int

match_lst = [(int(i),int(j)) for i,j in match_list]

products = [a * b for a,b in match_lst]

total=sum(products)
```

```python
total
```

```python
# part 2 want to skip any mul instruction coming after dont until a do is received

# regular expressions to find mul(a,b) in the string - 
# | - or condition to include do() and don't() as a third element in the tuple
# if a do() or dont is found it will be added in the third element of tuple like  ('', '', 'do()'),

p2_key = r"(?:mul\((\d+),(\d+)\))|(do\(\)|don't\(\))"


#

p2_match_list = re.findall(p2_key,dump)

# as per instructions - multiplications are valid at the start so start with valid as true
mul_valid = True

p2_total = 0

# for every tuple in the list
for i in p2_match_list:
    # check if the 3rd element of tuple is spaces and also check if we are still in valid case(ie - not followed by dont)
    if i[2] == '' and mul_valid:
        # if yes, make them as int and product the first two elements and add them to total
        p2_product = int(i[0]) * int(i[1])
        p2_total += p2_product
    # if 3rd element is not spaces or the mul_valid is false
    else:
        # then check if the 3rd element is a do - if so, then set valid flag as true for considering next mul
        if i[2] == 'do()':
            mul_valid = True
        # if the 3rd element is dont - then, set valid flag as false to skip the muls coming up
        else:
            mul_valid = False
            

print(p2_total)





```
