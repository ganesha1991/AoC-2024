```python
with open(r"C:\Users\thirggs\Downloads\Advent_of_code\2024\inputs\6.txt") as mapdump:
    # open the file, read entire file as string, strip the leading and trailing spaces, split based on lines into a list
    map_grid = mapdump.read().strip().split("\n")
    
# x and y axis are found by getting length of the list and length of an element in the list

rows = len(map_grid)
columns = len(map_grid[0])

print(rows,columns)
```

```python
print(map_grid)
```

```python
# first lets find the starting position of the security(^) in the grid
found = False

# start with found flag as false.. this found flag is required to stop the checking once we detected the security since it is 
# present only once. 

# checking for the whole list across each individual character 
for i in range(rows):
    for j in range(columns):
        if map_grid[i][j] == "^":
            # if found - store the location in a tuple and set found flag - then break out of the for loop(across columns)
            start_loc = (i,j)
            found = True
            break
            
    # this additional check and break is to skip the rest of the execution if found. 
    # without this even if the it comes out of first for loop - it will execute for the following rows
    # which is not required as the values are already found
    if found:
        break

    
            
print(start_loc) 
```

```python
# setup the directions for the security to move 
# security will move only 1 position forward only - and he turn right 90 degrees. if he meets an obstacle ('#')
# so, no diagonal movement - it will be either on one axis at a time not across both axis
# with starting position is facing up ^ initial movement will be on x axis towards up, 
# on obstacle turn right along y axis towards right 
# then another obstacle towards down(x axis) -
# then another obstacle(right) towards left (y axis)
# directions will be [-1,0] (up),[0,1](right), [1,0](down), [0,-1](left)

curr_direction = 0 # initially zero - as the ^ is facing up initally

# make initial position as starting location - assign variables to  position
current_pos = start_loc
x,y = current_pos

# direction matrix
dd = [[-1, 0], [0, 1], [1, 0], [0, -1]]

# initialise a set - for storing all the positions the security moves in the grid 
# using set instead of a list because we need to get only the unique positions - no duplicates
all_pos = set()

while True:
    # add the position into the positions set
    all_pos.add((x,y))
    
    # get the new movement offset from direction matrix based on direction. Based on direction - the offset will change
    # curr_Direction will be on range of length of list dd - so, it will be on one of the element in the list
    dx,dy = dd[curr_direction]
    
    # calculate the new co-ordinates after movement - add current position with offset
    next_x = x+dx
    next_y = y+dy
    
    # next_i = i + dd[dir][0]
    # next_j = j + dd[dir][1]
    
    # check if the new co-ordinates are within grid - if yes, move out of the loop (Security is out of grid) - if no,repeat
    if not (0 <= next_x < rows and 0 <= next_y < rows):
        print(next_x,next_y)
        break
        
    # check for the obstacles at the new co-ordinates
    if map_grid[next_x][next_y] == "#":
        # if obstacle, then change the direction to next direction in the direction matrix - 
        # to rotate across the direction matrix of length 4 - we are using modulus division of 4 
        curr_direction = (curr_direction + 1) % 4
    else:
        # if no obstacle, then change the curr postion to new position
        x,y = next_x,next_y
    
        
print(rows,columns)
print(len(all_pos))
```
