```python
with open(r"C:\Users\thirggs\Downloads\Advent_of_code\2024\inputs\9.txt") as filedump:
    # open the file, read entire file as string, strip the leading and trailing spaces, split based on newline character
    disk_map = filedump.read().strip()
```

```python
def file_processor(diskmap):
    processed_op = []
    
    file_block = True
    file_id = 0
    
    for i in diskmap:
        i = int(i)
        if file_block:
            processed_op += [file_id] * i
            file_id += 1
            file_block = False
        else:
            processed_op += ["."] * i
            file_block = True
            
    
    return processed_op


mapped_file_system = file_processor(disk_map)
```

```python
index=0
    
while mapped_file_system[index] != ".":
    index += 1
    
i = len(mapped_file_system) - 1
while mapped_file_system[i] == ".":
    i -= 1
        
while i > index:
    mapped_file_system[index] = mapped_file_system[i]
    mapped_file_system[i] = "."
    
    while mapped_file_system[index] != ".":
        index += 1
        
    while mapped_file_system[i] == ".":
        i -= 1
        
print(mapped_file_system)

    
    
```

```python
check_sum = 0

for i,j in enumerate(mapped_file_system):
    if j != ".":
        check_sum += i * j

print(check_sum)
```
